```python
# Import modules, data, variables
from datetime import datetime
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

path_daily = '/Users/carina/Desktop/code/NooksackWaterBudget/CurvyTemperature/'
```

```python
date_rng = pd.date_range(start='10/14/2015', end='9/4/18', freq='D')
date_rng
df = pd.DataFrame(date_rng, columns=['date'])
#df['data'] = np.random.randint(0,100,size=(len(date_rng)))
df['datetime'] = pd.to_datetime(df['date'])
df = df.set_index('datetime')
df.drop(['date'], axis=1, inplace=True)
```

```python
Lapse1_2017 = pd.read_csv(path_daily + 'Lapse1_data_2017_dailyRH.csv')
Lapse1_2018 = pd.read_csv(path_daily + 'Lapse1_data_2018_dailyRH.csv')
Lapse1_2017.columns = ['Time_Lapse1_2017_RHdaily', 'RH_Lapse1_daily_mean_2017']

Lapse1_2018.columns = ['Time_Lapse1_2018_RHdaily', 'RH_Lapse1_daily_mean_2018']

Lapse1_2017 = Lapse1_2017.set_index(pd.DatetimeIndex(Lapse1_2017['Time_Lapse1_2017_RHdaily']))

Lapse1_2018 = Lapse1_2018.set_index(pd.DatetimeIndex(Lapse1_2018['Time_Lapse1_2018_RHdaily']))

df_concat_Lapse1 = pd.concat([df, Lapse1_2017, Lapse1_2018], axis = 1,  join_axes = [df.index])
df_concat_Lapse1['RHdaily_lapse1'] = np.nan
#df_concat.AirT_Lapse2_daily_mean_2018.plot()
temp_1 = df_concat_Lapse1['RH_Lapse1_daily_mean_2017'].to_frame()
temp_2 = df_concat_Lapse1['RH_Lapse1_daily_mean_2018'].to_frame()
temp_1.columns = ['RHdaily']
temp_2.columns = ['RHdaily']

df_concat_Lapse1['RHdaily_lapse1'] = temp_1.combine_first(temp_2)
df_concat_Lapse1.RHdaily_lapse1.plot()
df_concat_Lapse1.head()
```

```python
Lapse5_2017 = pd.read_csv(path_daily + 'Lapse5_data_2017_dailyRH.csv')
Lapse5_2018 = pd.read_csv(path_daily + 'Lapse5_data_2018_dailyRH.csv')
Lapse5_2017.columns = ['Time_Lapse5_2017_RHdaily', 'RH_Lapse5_daily_mean_2017']

Lapse5_2018.columns = ['Time_Lapse5_2018_RHdaily', 'RH_Lapse5_daily_mean_2018']

Lapse5_2017 = Lapse5_2017.set_index(pd.DatetimeIndex(Lapse5_2017['Time_Lapse5_2017_RHdaily']))


Lapse5_2018 = Lapse5_2018.set_index(pd.DatetimeIndex(Lapse5_2018['Time_Lapse5_2018_RHdaily']))


df_concat_Lapse5 = pd.concat([df, Lapse5_2017, Lapse5_2018], axis = 1,  join_axes = [df.index])
df_concat_Lapse5['RHdaily_lapse1'] = np.nan
#df_concat.AirT_Lapse2_daily_mean_2018.plot()
temp_1 = df_concat_Lapse5['RH_Lapse5_daily_mean_2017'].to_frame()
temp_2 = df_concat_Lapse5['RH_Lapse5_daily_mean_2018'].to_frame()
temp_1.columns = ['RHdaily']
temp_2.columns = ['RHdaily']

df_concat_Lapse5['RHdaily_lapse5'] = temp_1.combine_first(temp_2)
df_concat_Lapse5.RHdaily_lapse5.plot()
df_concat_Lapse5.head()
```

```python
Lapse7_2016 = pd.read_csv(path_daily + 'Lapse7_data_2016_dailyRH.csv')
Lapse7_2017 = pd.read_csv(path_daily + 'Lapse7_data_2017_dailyRH.csv')
Lapse7_2018 = pd.read_csv(path_daily + 'Lapse7_data_2018_dailyRH.csv')

Lapse7_2016.columns = ['Time_Lapse7_2016_RHdaily', 'RH_Lapse7_daily_mean_2016']
Lapse7_2017.columns = ['Time_Lapse7_2017_RHdaily', 'RH_Lapse7_daily_mean_2017']
Lapse7_2018.columns = ['Time_Lapse7_2018_RHdaily', 'RH_Lapse7_daily_mean_2018']

Lapse7_2016 = Lapse7_2016.set_index(pd.DatetimeIndex(Lapse7_2016['Time_Lapse7_2016_RHdaily']))
Lapse7_2017 = Lapse7_2017.set_index(pd.DatetimeIndex(Lapse7_2017['Time_Lapse7_2017_RHdaily']))
Lapse7_2018 = Lapse7_2018.set_index(pd.DatetimeIndex(Lapse7_2018['Time_Lapse7_2018_RHdaily']))

df_concat_Lapse7 = pd.concat([df, Lapse7_2016, Lapse7_2017, Lapse7_2018], axis = 1,  join_axes = [df.index])
df_concat_Lapse7['RHdaily_lapse7'] = np.nan

temp_1 = df_concat_Lapse7['RH_Lapse7_daily_mean_2016'].to_frame()
temp_2 = df_concat_Lapse7['RH_Lapse7_daily_mean_2017'].to_frame()
temp_3 = df_concat_Lapse7['RH_Lapse7_daily_mean_2018'].to_frame()

temp_1.columns = ['RHdaily']
temp_2.columns = ['RHdaily']
temp_3.columns = ['RHdaily']

df_concat_Lapse7['RHdaily_lapse7'] = temp_1.combine_first(temp_2).combine_first(temp_3)
#df_concat_Lapse4['Tdaily_lapse1'] = temp_2.combine_first(temp_3)

df_concat_Lapse7.RHdaily_lapse7.plot()
df_concat_Lapse7.head()
```

```python
#combine all

df_all = df
df_all['RHdaily_Lapse1'] = df_concat_Lapse1['RHdaily_lapse1']
df_all['RHdaily_Lapse5'] = df_concat_Lapse5['RHdaily_lapse5']
df_all['RHdaily_Lapse7'] = df_concat_Lapse7['RHdaily_lapse7']

#a = df_all.values
#a = df_all.isnull().any().any()
#a  = a[~np.isnan(a)]
#df_all.combine(df_all, lambda s: s if isnan(s) else np.NaN)
#a
```
